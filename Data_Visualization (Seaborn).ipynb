```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('./nba.csv')
df.head()
```

```python

sns.displot(df['Age'], color='r', kde=False, bins=20)
```

```python
sns.lineplot(x='Age', y='Weight', data=df, color='r')
```

```python
sns.scatterplot(x='Age', y='Weight', data=df)
```

```python
sns.scatterplot(x='Age', y='Weight', data=df, hue=df['Position'])
```

```python
sns.boxplot(x='Age', y='Weight', data=df)
```

```python
sns.violinplot(x='Age', data=df)
```

```python
sns.countplot(x='Position', data=df)
```
